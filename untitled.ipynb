```python
from collections import deque

def water_jug_problem(capacity_a, capacity_b, target):
    visited = set()
    queue = deque([(0, 0)])
    visited.add((0, 0))

    print(f"Starting Water Jug Problem with capacities: A = {capacity_a}, B = {capacity_b}, Target = {target}\n")

    while queue:
        a, b = queue.popleft()
        print(f"Processing state: ({a}, {b})")

        if a == target or b == target:
            print("\nTarget reached!")
            return True

        possible_moves = []

        if (capacity_a, b) not in visited:
            possible_moves.append((capacity_a, b))

        if (a, capacity_b) not in visited:
            possible_moves.append((a, capacity_b))
        if (0, b) not in visited:
            possible_moves.append((0, b))

        if (a, 0) not in visited:
            possible_moves.append((a, 0))

        pour_to_b = min(a, capacity_b - b)
        if (a - pour_to_b, b + pour_to_b) not in visited:
            possible_moves.append((a - pour_to_b, b + pour_to_b))

        pour_to_a = min(b, capacity_a - a)
        if (a + pour_to_a, b - pour_to_a) not in visited:
            possible_moves.append((a + pour_to_a, b - pour_to_a))

        for new_state in possible_moves:
            visited.add(new_state)
            queue.append(new_state)
            print(f"  -> Adding new state: {new_state}")

        print(f"Queue after processing ({a}, {b}): {list(queue)}\n")

    print("Target cannot be reached.")
    return False

capacity_a = 10
capacity_b = 3
target = 8

result = water_jug_problem(capacity_a, capacity_b, target)
print("\nResult:", result)
```
